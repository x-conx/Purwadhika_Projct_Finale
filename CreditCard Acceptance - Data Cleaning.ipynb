```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
import seaborn as sns
%matplotlib inline
```

# read csv file

```python
df = pd.read_csv('creditcardmarketing-bbm.csv')
```

```python
df
```

# cleaning null data

```python
df[df.isna().any(axis=1)]
```

```python
df[df.isna().any(axis=1)].count()
```

```python
df.dropna(axis=0,inplace=True)
```

```python
df
```

```python
df.drop(['Customer Number'],axis=1,inplace=True)
```

```python
df
```

# change target from yes/no to 1/0

```python
new_int = []
for i in df['Offer Accepted']:
    if i == 'No':
        new_int.append(0)
    else:
        new_int.append(1)
df['Offer Accepted']=new_int
df
```

# change column name(original is hard to work with)

```python
df.columns
```

```python
df.columns = ['Offer_Accepted', 'Reward', 'Mailer_Type', 'Income_Level',
       'Bank_Accnt_Open', 'Overdraft_Protection', 'Credit_Rating',
       'Credit_Cards_Held', 'Homes_Owned', 'Household_Size',
       'Own_Your_Home', 'Average_Balance', 'Q1_Balance', 'Q2_Balance',
       'Q3_Balance', 'Q4_Balance']
```

```python
df
```

# change credit rating

```python
df['Credit_Rating'].unique()
```

```python
listx = []
for i in df['Credit_Rating']:
    if i == 'High':
        listx.append(3)
    elif i=='Medium':
        listx.append(2)
    else:
        listx.append(1)
df['Credit_Rating']=listx
df
```

# export clean dataset to csv file

```python
df.to_csv('CreditCardDataset_cleaned.csv',index=False)
```
